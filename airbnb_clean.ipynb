# Preparing AirBnB data

## Read File & Specify Datatypes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
file_path = '/Applications/PostgreSQL 17/Raw CSVs/Airbnb_Open_Data.csv'

# Specifying the data types of the columns because many of the columns include multiple data types which can cause errors in reading the data
df = pd.read_csv(file_path, dtype ={
    'id': int,
    'name': str,
    'host_id': float,
    'host_identity_verified': str,
    'host_name': str,
    'neighbourhood_group': str,
    'neighbourhood': str,
    'lat': float,
    'long': float,
    'country': str,
    'country code': str,
    'instant_bookable': str,
    'room_type': str,
    'construction_year': float,
    'price': str,  
    'service_fee': str,  
    'minimum_nights': float,
    'number_of_reviews': float,
    'last_review': str,
    'reviews_per_month': float,
    'review_rate_number': float,
    'calculated_host_listings_count': float,
    'availability_365': float,
    'house_rules': str,
    'license': str
})

print(df.dtypes) # check datatypes

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)

# Enable inline plotting
%matplotlib inline
```

## Remove duplicate rows

```python
df = df.drop_duplicates()

df
```

## Remove unnecessary column(s)

```python
df = df.drop(columns = ['license','house_rules','country code','id','host id'])

df
```

## Cleaning data entry errors

## Remove inconsistencies/errors in Data Validity

### Unverified hosts

```python
# Airbnb does not allow the listings of properties on their website without the identity of the host being verified or the host having at least part of their name displayed

df = df[~df['host_identity_verified'].isin(['', 'unconfirmed']) & df['host_identity_verified'].notna()] # Filters data to exclude all rows where cells in host_identity_verified are blank or read 'unconfirmed'

df = df[df['host name'].notna()] # Filters out all rows where no host name is displayed with the listing

df
```

### Missing listing names

```python
# Airbnb listings are required to include a name for the listing (i.e. 'Skylit Midtown Castle')

df = df[df['NAME'].notna()] # Filters out all rows where no name is displayed with the listing

df
```

### Missing price data

```python
# Airbnb requires a price be listed

df = df[df['price'].notna()] # This code removes all rows where the price is not listed

df
```

### Convert numeric data stored as object to float for any potential analyses

```python

df['price'] = pd.to_numeric(df['price'].replace({'\$': '', ',': ''}, regex=True).str.strip(), errors='coerce')  # 'price' column is converted to numeric and removes the '$' sign, commas, and any leading/trailing spaces

df['service fee'] = pd.to_numeric(df['service fee'].replace({'\$': '', ',': ''}, regex=True).str.strip(), errors='coerce')  # 'service fee' column is converted to numeric and removes the '$' sign, commas, and any leading/trailing space

df['price'] = df['price'].astype(float) # Converts the 'price' column to float

# Check the data types again to ensure both are floats
print(df['price'].dtype)
print(df['service fee'].dtype)
print(df['minimum nights'].dtype)

# Syntax warning here can be ignored
```

### Impossibilities with numerical data

```python
# Filter out rows with negative values where negative values would not be possible
# The data types must be floats for this to work

df = df[(df['price'] >= 0)]
df = df[(df['service fee'] >= 0)]
df


```

### Reset Index

```python
df = df.reset_index(drop=True)
df
```

## Analysis & Plots

### Minimum Nights

```python
night_df = df[['minimum nights']].copy()
night_df['minimum nights'].plot(kind = 'box')
```

```python
night_df['minimum nights'] = night_df['minimum nights'].fillna(night_df['minimum nights'].median())
```

```python
night_df['minimum nights'].describe()
```

When calculating IQR, lowerbound, and upperbound for minimum nights, we see that the lowerbound is negative. This is not possible, so we will remove the negative values by clipping the data.

These are the calculation steps for IQR, lowerbound, and upperbound and to analyze the minimum nights data with a box plot.

IQR = Q3 - Q1

Lower = Q1 - 1.5*IQR

Upper = Q3 - 1.5*IQR

q1 = np.percentile(data, 25)

q3 = np.percentile(data, 75)

minimum = np.min(data)

median = np.median(data)

maximum = np.max(data)

```python
q1 = np.percentile(night_df['minimum nights'], 25)
median = np.median(night_df['minimum nights'])
q3 = np.percentile(night_df['minimum nights'], 75)
IQR = q3 - q1
lower_min = q1 - 1.5*IQR
upper_min = q3 + 1.5*IQR
print('Lower Min:',lower_min)
print('Upper Min:',upper_min)
print('IQR:',IQR)
```

```python
night_df['minimum nights'].clip(lower=0, upper=30, inplace=True)

night_df['minimum nights'].plot(kind = 'box')
```

```python
night_df['minimum nights'].plot(kind='hist', bins=20, edgecolor='black')
plt.title('Histogram of Minimum Nights')
plt.xlabel('Minimum Nights')
plt.ylabel('Frequency')
plt.gca().spines['top'].set_visible(False)
plt.show()
```

### Price by Neighbourhood

```python
df.groupby('neighbourhood group')['price'].mean().plot(kind = 'bar', edgecolor='black')
plt.title('Average Price per Night by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price per Night')
plt.gca().spines['top'].set_visible(False)
plt.show()
```

### Price by Room Type

```python
df.groupby('room type')['price'].mean().plot(kind = 'bar')
plt.title('Average Price per Night by Room Type', fontsize = 12)
plt.ylabel('Average Price per Night', fontsize = 10)
plt.xlabel('Room Type', fontsize = 10)
plt.gca().spines['top'].set_visible(False)
plt.xticks(rotation=0,fontsize = 10) 
```

### Neighborhoods with the most listings

```python
neighbourhood_listings = df['neighbourhood'].value_counts().head(10)


axis_labels = neighbourhood_listings.plot(kind = 'bar', figsize=(14,8), edgecolor = 'black', color = 'seagreen') 
axis_labels.bar_label(axis_labels.containers[0], label_type = 'edge') 
plt.title('Top 10 Neighbourhoods by Number of Listings\n',loc = 'left',font = 'Times New Roman', fontsize = 20)
plt.xlabel('Neighbourhood', font = 'Times New Roman',fontsize = 15)
plt.ylabel('Number of Listings', font = 'Times New Roman',fontsize = 15)
plt.xticks(rotation=45, font = 'Times New Roman',fontsize = 12)
plt.gca().spines['top'].set_visible(False)
plt.gca().set_facecolor((0.83, 0.83, 0.83, 0.3))
plt.show()


```

### Average Price by Construction year  

```python
# Converting Construction year to an int to correctly display years
cy_df = df.dropna(subset=['Construction year']).copy()
cy_df['Construction year'] = cy_df['Construction year'].astype(int)
```

```python
avg_construct = cy_df.groupby('Construction year')['price'].mean().reset_index()

plt.figure(figsize=(14,8))
plt.scatter(avg_construct['Construction year'], avg_construct['price'], color='royalblue')
plt.title('Average Price per Night by Construction Year')
plt.xlabel('Construction Year')
plt.ylabel('Average Price per Night')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
z = np.polyfit(avg_construct['Construction year'], avg_construct['price'], 1)
p = np.poly1d(z)
plt.plot(avg_construct['Construction year'], p(avg_construct['Construction year']), color='crimson')

plt.show()
```

```python
cy_df.groupby('Construction year')['price'].mean().plot(kind = 'line')
plt.title('Average Price per Night by Construction Year',fontsize = 12)
plt.xlabel('Construction Year',fontsize = 10)
plt.ylabel('Average Price per Night',fontsize = 10)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.show()
```
